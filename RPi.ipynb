```python
import RPi.GPIO as GPIO
import time

# Disable warnings and clean up GPIO
GPIO.setwarnings(False)
GPIO.cleanup()

# Set the GPIO mode to BCM
GPIO.setmode(GPIO.BCM)

# Define the GPIO pins
TRIG = 4
ECHO = 18
GREEN = 17
YELLOW = 27
RED = 22

# Set up the GPIO pins
GPIO.setup(TRIG, GPIO.OUT)
GPIO.setup(ECHO, GPIO.IN)
GPIO.setup(GREEN, GPIO.OUT)
GPIO.setup(YELLOW, GPIO.OUT)
GPIO.setup(RED, GPIO.OUT)

# Function to turn on the green light
def green_light():
    GPIO.output(GREEN, GPIO.HIGH)
    GPIO.output(YELLOW, GPIO.LOW)
    GPIO.output(RED, GPIO.LOW)

# Function to turn on the yellow light
def yellow_light():
    GPIO.output(GREEN, GPIO.LOW)
    GPIO.output(YELLOW, GPIO.HIGH)
    GPIO.output(RED, GPIO.LOW)

# Function to turn on the red light
def red_light():
    GPIO.output(GREEN, GPIO.LOW)
    GPIO.output(YELLOW, GPIO.LOW)
    GPIO.output(RED, GPIO.HIGH)

# Function to measure the distance using the ultrasonic sensor
def get_distance():
    # Trigger the sensor
    GPIO.output(TRIG, True)
    time.sleep(0.00001)
    GPIO.output(TRIG, False)

    # Measure the time between trigger and echo
    while GPIO.input(ECHO) == False:
        start = time.time()

    while GPIO.input(ECHO) == True:
        end = time.time()

    # Calculate the time it took for the echo to return
    signal_time = end - start

    # Convert the signal time to distance in cm (speed of sound = 34300 cm/s)
    distance = signal_time * 34300 / 2
    return distance

# Main loop to continuously check the distance and control lights
try:
    while True:
        distance = get_distance()
        print(f"Distance: {distance:.2f} cm")
        
        # Control the traffic lights based on the distance
        if distance >= 25:
            green_light()
        elif 10 < distance < 25:
            yellow_light()
        elif distance <= 10:
            red_light()

        time.sleep(0.05)

except KeyboardInterrupt:
    GPIO.cleanup()
```
